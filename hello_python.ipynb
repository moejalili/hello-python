```python
class Person(object):
    def __init__(self, first, last):
        self.first_name = first
        self.last_name = last
    def comma_name(self):
        return '{0}, {1}'.format(self.last_name, self.first_name)
```

```python
p = Person('Moe', 'Jalili')
```

```python
p.comma_name()
```

### Managing Group of Cells

```python
def sieve(n):
    not_prime = []
    primes = []
    for i in range(2, n+1):
        if i not in not_prime:
            primes.append(i)
            for j in range(i * i, n + 1, i):
                not_prime.append(j)
    return primes
```

```python
sieve(50)
```
